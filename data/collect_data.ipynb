```python
## Columns to Keep
columns_to_keep = [
    'id',          # unique person node
    'name',        # person name
    'aliases',     # other names
    'birth_date',  # temporal attribute
    'countries',   # nationality / wanted by
    'sanctions',   # type of crime
    'first_seen',  # timeline
    'last_seen',   # timeline
    'last_change'  # history / updates
]
Interpol_data = Interpol_data[columns_to_keep] 
Interpol_data
```

```python
print("=== Before Imputation ===")
print(Interpol_data.isnull().sum())
```

```python
import pandas as pd

# For aliases - fill with empty list or string
Interpol_data['aliases'] = Interpol_data['aliases'].fillna('0')  # or .fillna([])

# For birth_date - keep null but create indicator
Interpol_data['birth_date_missing'] = Interpol_data['birth_date'].isna().astype(int)

# For countries - fill with 'Unknown'
Interpol_data['countries'] = Interpol_data['countries'].fillna('Unknown')

# For sanctions - verify then fill
Interpol_data['sanctions'] = Interpol_data['sanctions'].fillna('No')  # or verify first

# Alternative: Drop rows only if multiple critical fields are missing
df_cleaned = Interpol_data.dropna(subset=['birth_date', 'countries', 'sanctions'], 
                       thresh=2)  # keep if at least 2 of 3 are present
```

```python
print("=== After Imputation ===")
print(Interpol_data.isnull().sum())
```

```python
Interpol_data.rename(columns=lambda x: x.strip(), inplace=True)  # Remove leading/trailing spaces from column names
```
